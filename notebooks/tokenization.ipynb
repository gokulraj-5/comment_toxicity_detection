```python
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import train_test_split

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
train_df = pd.read_csv("../data/processed/cleaned_train.csv")

train_df.head()
```

```python
target_columns = [
    "toxic",
    "severe_toxic",
    "obscene",
    "threat",
    "insult",
    "identity_hate"
]
```

```python
X = train_df["clean_comment"]

y = train_df[target_columns]
```

```python
print(X.shape)
print(y.shape)
```

```python
train_df["clean_comment"] = (
    train_df["clean_comment"]
    .fillna("")
    .astype(str)
)
```

```python
train_df["clean_comment"].isna().sum()
```

```python
train_df["clean_comment"].dtype
```

```python
X = train_df["clean_comment"]

print(type(X.iloc[0]))
```

```python
MAX_WORDS = 10000

tokenizer = Tokenizer(
    num_words=MAX_WORDS,
    oov_token="<OOV>"
)
```

```python
tokenizer.fit_on_texts(X)
```

```python
train_df[train_df["clean_comment"].isna()]
```

```python
with open("../artifacts/tokenizer.pkl", "wb") as file:
    pickle.dump(tokenizer, file)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(X)
```

```python
print(sequences[0])
```

```python
MAX_LEN = 200
```

```python
X_padded = pad_sequences(
    sequences,
    maxlen=MAX_LEN,
    padding="post",
    truncating="post"
)
```

```python
X_padded.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_padded,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_val.shape)
```

```python
np.save("../artifacts/X_train.npy", X_train)
np.save("../artifacts/X_val.npy", X_val)

np.save("../artifacts/y_train.npy", y_train)
np.save("../artifacts/y_val.npy", y_val)
```
